```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 트랜잭션 구성
dataset = [['겨울왕국', '미녀와 약수', '쥬라기 월드', '아바타', '타이타닉'],
           ['스타워즈', '어벤져스', '아이언맨'],
           ['스타워즈', '아바타', '라이온 킹', '어벤져스'],
           ['쥬라기 월드', '아바타', '해리포터'],
           ['스타워즈', '겨울왕국', '미녀와 야수', '라이온 킹', '타이타닉'],
           ['스타워즈', '어벤져스', '분노의 질주', '쥬라기 월드']]
```

```python
# 리스트의 리스트로된 트랜젹션 데이터를 Numpy 배열로 인코딩
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# 지지도가 0.3 이상인 값만 확인
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets
```

```python
# 신뢰도가 0.3 이상인 값만 확인
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3, num_itemsets=len(frequent_itemsets))
```

```python
# 향상도 확인
df_assoc = association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=len(frequent_itemsets))
df_assoc.sort_values('lift', ascending=False)
```

### 제시하는 추천 목록
수업에 나오는대로 신뢰도, 지지도, 향상도 순으로 분석을 하였다.
마지막 결과값을 통해서도 알 수 있듯이 데이터의 값이 적어 방향성이 같은 것을 알 수 있다.
타이타닉을 좋아하면 겨울왕국도 좋아하고,겨울왕국을 좋아하면 타이타닉을 좋아하는것이다. 
이 결과를 기반으로 추천목록을 만든다면 [(타이타닉, 겨울왕국), (라이온 킹, 스타워즈), (어벤져스, 스타워즈), (아바타, 쥬라기 월드)] 이렇게 묶어서 추천할것 같다.
