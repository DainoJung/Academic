```python
import pandas as pd
import numpy as np
```

```python
# 데이터 로딩
data = pd.read_csv('./data/heights.csv')
data
```

```python
data.shape
```

```python
# 결측치 확인
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt

X = data["father"]
y = data["son"]

plt.plot(X, y, '.')
plt.xlabel('Father')
plt.ylabel('Son')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.values
```

```python
# Train data(Blue) and Test data(Red) 분포 파악
plt.plot(X_train, y_train, 'b.', label='train data')
plt.plot(X_test, y_test, 'r.', label='test data')
plt.legend()
plt.xlabel('X train')
plt.ylabel('y train')
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# reshape(-1, 1)으로 컬럼 갯수가 1개인 행렬로 조정
lr.fit(X_train.values.reshape(-1,1), y_train)
```

```python
# 회귀계수(기울기와 절편)
[lr.coef_, lr.intercept_]
```

```python
# train 데이터의 x값 범위(최소값, 최대값)
x = np.array([min(X_train), max(X_train)])
x
```

### 회귀모형을 통한 예

```python
# 예측함수만들기 : y = b(절편) + a(기울기)x
f = lambda x: lr.intercept_ + (lr.coef_ * x)
```

```python
# f 함수를 이용
f(180)
```

### 시각화

```python
# train 데이터의 분포
plt.plot(X_train, y_train, 'b.')
plt.xlabel('X train')
plt.ylabel('y train')
plt.plot(x, f(x), 'r')
```

### 성능평가

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_pred = lr.predict(X_test.values.reshape(-1,1))
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
```

```python
# 결정계수
r2 = r2_score(y_test, y_pred)
```

```python
print("MAE : %.2f" %mae)
print("MSE : %.2f" %mse)
print("RMSE : %.2f" %rmse)
print("R2 : %.2f" %r2)
```
