## Bar Plot

```python
import numpy as np
import pandas as pd
blood=np.array(['A', 'O', 'B', 'B', 'O', 'AB', 'A', 'O'])
blood
```

```python
bloodtype=pd.Series(blood).value_counts()
bloodtype
```

```python
bloodtype.plot(kind='bar')
```

```python
s=pd.Series(np.random.rand(10))
s
```

```python
s.plot(kind='bar')
```

```python
df=pd.DataFrame(np.random.rand(10, 4),
               columns=['a', 'b', 'c', 'd'])
df
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='barh', stacked=True)
```

```python
alco2009=pd.read_csv('niaaa-report2009.csv', index_col='State')
alco2009.mean().plot(kind='bar')
```
