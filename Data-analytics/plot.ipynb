```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
alco=pd.read_csv('Data/niaaa-report.csv', index_col=['State','Year'])
alco.groupby('State').mean().plot(kind='bar')
```

```python
alco.hist()
```

```python
plt.hist(alco['Beer'], label='Beer')
plt.hist(alco['Wine'], label='Wine')
plt.legend(loc='upper right')
```

### Kernel Density Estimation Chart

```python
alco.plot(kind='kde')
plt.xlim(0, 2)  # x  0 2.5 
plt.show()
```

```python
alco.loc['California'].plot(kind='kde')
plt.show()
```
