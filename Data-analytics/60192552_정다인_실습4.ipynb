```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 데이터 불러오기
data = pd.read_csv('Data/niaaa-report.csv', index_col=['State', 'Year'])
```

## 1번 문제

```python
# State = California
# Year = 1990 ~ 1999
cali90 = data.loc[('California', slice(1990, 1999)), :].copy()
cali90
```

## 2번 문제

```python
# State = New York
# Year = 1990 ~ 1999
ny90 = data.loc[('New York', slice(1990, 1999)), :].copy()
ny90
```

## 3번 문제

```python
cali90.loc[('California', 1999), 'Beer'] = np.nan
cali90
```

```python
ny90.loc[('New York', 1991), 'Wine'] = np.nan
ny90
```

## 4번 문제

```python
cali_ny_90 = pd.concat([cali90, ny90])
cali_ny_90
```

## 5번 문제

```python
cali_ny_90.dropna(axis=0, how='any')
```

## 6번 문제

### 주 별 알코올 소비량의 평균과 최대, 최솟값

```python
# Beer, Wine, Spirits의 합으로 All_Alcohol 열 추가
cali_ny_90['All_Alcohol'] = cali_ny_90[['Beer', 'Wine', 'Spirits']].sum(axis=1)

# State별 All_Alcohol의 평균, 최대, 최소값 계산
cali_ny_all_stats = cali_ny_90.groupby('State')['All_Alcohol'].agg(['mean', 'max', 'min'])

# California의 평균 알코올 소비량이 더 많음
cali_ny_all_stats
```

### 주종별 각 주의 최대, 최소 소비량

```python
# 주종별 각 주의 최대 및 최소 소비량 계산
cali_ny_max_min = cali_ny_90.groupby('State')[['Beer', 'Wine', 'Spirits']].agg(['max', 'min'])

# 결과 출력
print("California and New York Alcohol Consumption (Max and Min) by Type (1990-1999):")
print(cali_ny_max_min)
```
